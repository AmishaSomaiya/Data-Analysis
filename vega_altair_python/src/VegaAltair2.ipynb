```python
import pandas as pd
import altair as alt
from statistics import mean
 
```

```python
url = "C:\\Users\\hrsom\\Downloads\\sunshine.csv"
source = pd.read_csv(url)

url1 = "C:\\Users\\hrsom\\Downloads\\sunshine_an.csv"
statis = pd.read_csv(url1)
```

```python
alt.Chart(source).mark_rect().encode(
    y=alt.Y('city:N', axis=alt.Axis(title='City')),
    x=alt.X('month:N', axis=alt.Axis(title='Month')),
    color=alt.Color('mean(sunshine):Q', scale=alt.Scale(scheme='yellowgreenblue'), legend=alt.Legend(title='Mean Sunshine (hours)'))
).properties(
    width=400,
    height=200,
    title='Mean Monthly Hours of Sunshine for Cities sorted by Distance to Equator(Latitude)'
)
```

```python
# Create a dataframe with the given data
df = pd.DataFrame({
    'city': ['Chicago', 'Houston', 'Miami', 'New York', 'San Francisco', 'Seattle'],
    'min': [106, 141, 209, 154, 156, 52],
    'max': [318, 298, 280, 302, 330, 312],
    'avg': [208, 219, 241, 223, 245, 180]
})

# Melt the dataframe to create a long-form dataframe
df_melt = pd.melt(df, id_vars=['city'], value_vars=['min', 'max', 'avg'], var_name='temp')

# Create the chart
chart = alt.Chart(df_melt).mark_circle(size=60).encode(
    x='city:N',
    y=alt.Y('value:Q', scale=alt.Scale(domain=[0, 350, 50]), axis=alt.Axis(title='Temperature (F)')),
    color=alt.Color('temp:N', scale=alt.Scale(domain=['min', 'max', 'avg'], range=['blue', 'red', 'orange']), 
                    legend=alt.Legend(title='Temperature'))
).properties(
    width=400,
    height=200,
    title='Temperature Ranges for Selected Cities'
)

# Display the chart
chart
```
