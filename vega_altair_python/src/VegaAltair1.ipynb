```python
import altair as alt
import pandas as pd

# read the data from the CSV file
url = "C:\\Users\\hrsom\\Downloads\\sunshine.csv"
# dataset = pd.read_csv(url)
data = pd.read_csv(url)

# define the base chart
base = alt.Chart(data).transform_calculate(
    year='2000+datum.monthnum'
).encode(
    x=alt.X('year:O', axis=alt.Axis(labelAngle=-45)),
    y=alt.Y('sunshine:Q', stack='normalize'),
    color='city:N'
)

# define the density chart
density = base.mark_area(opacity=0.3).encode(
    alt.X('sunshine:Q', bin=alt.Bin(maxbins=30)),
    alt.Y('density:Q')
)

# define the line chart
line = base.mark_line().encode(
    alt.Y('mean(sunshine):Q', axis=alt.Axis(format='%')),
    alt.Color('city:N', legend=None)
)

# combine the charts and set the chart title
chart = (density + line).properties(
    title='Stacked Density Estimation of Monthly Sunshine Percentage by City'
)

# display the chart
chart
```

```python
def cal_stats(lst):
    minval = min(lst)
    maxval = max(lst)
    avgval = mean(lst)
    return [minval, maxval, avgval]
```

```python
import pandas as pd
import altair as alt

# Create sample data
statis = pd.DataFrame({
    'city': ['Chicago', 'Houston', 'Miami', 'New York', 'SFO', 'Seattle'],
    'min': [106, 141, 209, 154, 156, 52],
    'max': [318, 298, 280, 302, 330, 312],
    'avg': [208, 219, 241, 223, 245, 180]
})

# Melt data to long format
statis_melted = pd.melt(statis, id_vars=['city'], var_name='Measure', value_name='Value')

# Sort measures by desired order
measure_order = ['min', 'avg', 'max']
statis_melted['Measure'] = pd.Categorical(statis_melted['Measure'], categories=measure_order, ordered=True)
statis_melted = statis_melted.sort_values(['city', 'Measure'])

# Create chart
alt.Chart(statis_melted).mark_circle(size=60).encode(
    x=alt.X('Measure:N', axis=alt.Axis(title='Measure')),
    y=alt.Y('Value:Q', axis=alt.Axis(title='Sunshine', tickCount=8)),
    color=alt.Color('city:N', legend=alt.Legend(title='City'))
).properties(
    width=400,
    height=200,
    title='Sunshine Statistics by City'
)
```
